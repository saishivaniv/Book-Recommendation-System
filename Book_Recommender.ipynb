Popularity Based Recommender System

```python
ratings_with_name = ratings.merge(books, on = 'ISBN')
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'}, inplace=True)
```

```python
num_rating_df.sample(3)
```

```python
#convert 'Book-Rating' column to numeric type
ratings_with_name['Book-Rating'] = pd.to_numeric(ratings_with_name['Book-Rating'],errors='coerce')
```

```python
#group by 'Book-Title' and calculate mean of 'Book-Rating'
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
```

```python
#Rename 'Book-Rating' to 'avg_ratings'
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'}, inplace=True)
```

```python
avg_rating_df.head(5)
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df.head(5)
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings', ascending=False).head(50)
```

```python
popular_df
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','num_ratings','Image-URL-M','avg_ratings']]
```

Collaborative Filtering Based Recommender System

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
padhe_likhe_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_rating = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
def recommend(book_name):
    # Check if book_name exists in pt.index
    if book_name not in pt.index:
        print(f"Book '{book_name}' not found in the dataset.")
        return []

    # Index fetch
    index = np.where(pt.index == book_name)[0][0]
    
    # Ensure similarity_scores has the correct shape
    if index >= len(similarity_score):
        print("Index out of range in similarity_scores.")
        return []

    # Fetch similar items
    similar_items = sorted(list(enumerate(similarity_score[index])), key=lambda x: x[1], reverse=True)[1:6]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        
        # Ensure temp_df is not empty
        if temp_df.empty:
            print(f"No data found for book: {pt.index[i[0]]}")
            continue
        
        # Deduplicate and extract information
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```
